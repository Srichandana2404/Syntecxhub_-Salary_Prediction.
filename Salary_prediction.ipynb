```python
#step1: importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv('SalaryPrediction.csv')
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().value_counts()
```

```python
plt.boxplot(df['Experience Years'])
plt.show()
```

```python
plt.scatter(df['Experience Years'],df['Salary'],color='purple',marker='+')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.title('Experience Years vs Salary')
plt.show()
```

```python
X=df[['Experience Years']]
y=df['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
model=LinearRegression()
model.fit(X_train, y_train)
```

```python
plt.scatter(X_train, y_train, color='blue', label='YofExp_sal_data')
plt.plot(X_train, model.predict(X_train), color='green', label='Regression Line')
plt.title('Year of Exp VS Sal')
plt.xlabel('Year of Exp')
plt.ylabel('Salary')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.scatter(X_test, y_test, color='blue', label='Test Data')

# Plot the regression line (same as the one for training)
plt.plot(X_train, model.predict(X_train), color='red', label='Regression Line')
plt.title('Test Data: year of exp VS Salary')
plt.xlabel('year of exp')
plt.ylabel('Salary')
plt.legend()
plt.grid()
plt.show()
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f'Mean Squared Error (MSE): {mse:.2f}')
print(f'R-Squared (R): {r2:.2f}')
```
